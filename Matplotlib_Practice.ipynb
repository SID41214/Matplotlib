```python
import matplotlib.pyplot as plt

data1 = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
data2 = [1, 3, 3, 3, 4, 4, 4, 5, 5, 6]

plt.hist([data1, data2], bins=5, label=['Data1', 'Data2'])
plt.title('Histogram with Multiple Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
