 1. **Multiple Line Plot with Different Line Styles and Widths**

```python
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4]
y1 = [1, 2, 3, 4, 5]
y2 = [2, 3, 4, 5, 6]
y3 = [3, 4, 5, 6, 7]

plt.plot(x, y1, linestyle='--', linewidth=1.5, label="Dashed Line")
plt.plot(x, y2, linestyle=':', linewidth=2.5, label="Dotted Line")
plt.plot(x, y3, linestyle='-', linewidth=3, label="Solid Line")
plt.grid(True)
plt.legend()
plt.show()
```

2. **Scatter Plot with Custom Marker Styles and Colors**

```python
x = [1, 2, 3]
y = [4, 5, 6]
plt.scatter(x, y, color='red', marker='o', s=100, label="Circle")
plt.scatter([2, 3, 4], [5, 6, 7], color='blue', marker='s', s=100, label="Square")
plt.scatter([3, 4, 5], [6, 7, 8], color='green', marker='^', s=100, label="Triangle")
plt.legend()
plt.show()
```

3. **Bar Graph with Edge Color and Face Color**

```python
x = ['A', 'B', 'C', 'D', 'E']
y = [5, 7, 8, 2, 6]
plt.bar(x, y, edgecolor='black', color='lightblue')
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

4. **Horizontal Bar Graph with Annotations**

```python
x = ['A', 'B', 'C', 'D', 'E']
y = [5, 7, 8, 2, 6]
bars = plt.barh(x, y, color=['red', 'blue', 'green', 'purple', 'orange'])

for bar in bars:
    plt.text(bar.get_width() + 0.1, bar.get_y() + 0.5, str(bar.get_width()), va='center')
plt.show()
```

5. **Pie Chart with Start Angle, Explode, and Shadow**

```python
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)

plt.pie(sizes, explode=explode, labels=labels, startangle=90, shadow=True, autopct='%1.1f%%')
plt.show()
```

6. **Subplots: Line Plot and Bar Graph**

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Line plot
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
ax1.plot(x, y, marker='o', color='b')
ax1.set_title("Line Plot")

# Bar graph
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 8, 2]
ax2.bar(categories, values, color='orange')
ax2.set_title("Bar Graph")

plt.show()
```

7. **Line Chart with Annotations and Legend**

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 9]
y2 = [1, 3, 4, 6, 8]

plt.plot(x, y1, label="Line 1", color='red')
plt.plot(x, y2, label="Line 2", color='blue')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

plt.annotate("Peak Value", xy=(4, 7), xytext=(5, 8),
             arrowprops=dict(arrowstyle="->", color='black'))
plt.show()
```

8. **Fill Area Between Two Lines**

```python
x = [0, 1, 2, 3, 4]
y1 = [1, 2, 3, 4, 5]
y2 = [2, 3, 4, 5, 6]

plt.plot(x, y1, color='blue')
plt.plot(x, y2, color='green')
plt.fill_between(x, y1, y2, color='gray', alpha=0.5)
plt.xlabel("X-Values")
plt.ylabel("Y-Values")
plt.show()
```

**9. Multiple Bar Plot (Grouped Bar Plot)**

```python
import numpy as np

categories = ['Group 1', 'Group 2', 'Group 3', 'Group 4', 'Group 5']
values1 = [5, 7, 8, 2, 6]
values2 = [3, 6, 9, 4, 5]
bar_width = 0.35
x = np.arange(len(categories))

plt.bar(x, values1, width=bar_width, label='Category A', color='blue')
plt.bar(x + bar_width, values2, width=bar_width, label='Category B', color='green')
plt.xlabel("Groups")
plt.ylabel("Values")
plt.legend()
plt.xticks(x + bar_width / 2, categories)
plt.show()
```

**10. Line Plot with Customized Gridlines and Line Markers**

```python
x = list(range(10))
y = [i ** 2 for i in x]

plt.plot(x, y, marker='o', markersize=8, markerfacecolor='red', linestyle='-.', color='blue')
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.show()
```
