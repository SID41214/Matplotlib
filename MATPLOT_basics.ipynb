```python
import numpy as np
import matplotlib.pyplot as plt
x= np.array(['a','b','c','d'])
y= np.array([3,8,7,4])
plt.bar(x,y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x= np.array(['a','b','c','d'])
y= np.array([3,8,7,4])
plt.bar(x,y,color=['r','g','y','b'],edgecolor='k',width=2,lw=3)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x= np.array(['a','b','c','d'])
y= np.array([3,8,7,4])
plt.barh(x,y,color=['r','w','y','b'],edgecolor='k',lw=4,height=2)#height is to increase the width of the horizontal bar
plt.grid()
plt.show()
```

###histogram

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.random.normal(170,10,250)
plt.hist(x)
plt.show()
```

###pie chart

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([35,25,25,15])
plt.pie(x,colors=['r','g','m','b'])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([35,25,25,15])
labels=['manga','thenga','eethapayam','payam']
plt.pie(x,colors=['r','g','m','b'],labels=labels)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([35,25,25,15])
labels=['manga','thenga','eethapayam','payam']
plt.pie(x,colors=['r','g','m','b'],labels=labels,startangle=0)
plt.show()
```

###explode

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([35,25,25,15])
labels=['manga','thenga','eethapayam','payam']
explo = [0,0,0.4,0]
plt.pie(x,colors=['r','g','m','b'],labels=labels,startangle=0,explode=explo)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([35,25,25,15])
labels=['manga','thenga','eethapayam','payam']
plt.pie(x,colors=['r','g','m','b'],labels=labels,startangle=0,shadow=True,explode=[0.4,4,1,20])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([35,25,25,15])
labels=['manga','thenga','eethapayam','payam']
plt.pie(x,colors=['r','g','m','b'],labels=labels,startangle=0,shadow=True,explode=[0.1,0.2,0.4,0.9])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([35,25,25,15])
labels=['manga','thenga','eethapayam','payam']
plt.pie(x,colors=['r','g','m','b'],labels=labels,startangle=0)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([35,25,25,15])
labels=['manga','thenga','eethapayam','payam']
plt.pie(x,colors=['r','g','m','b'],labels=labels,startangle=0)
plt.legend(title="mine")
plt.show()
```
