
1.	Create a simple line plot with the following y-coordinates: [1, 4, 9, 16, 25]

7.	Create a pie chart with the following data: sizes = [15, 30, 45, 10] and labels = ['A', 'B', 'C', 'D'].

---



```python
import matplotlib.pyplot as plt

sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

8.	Create a 2x1 subplot where the first plot is a line plot and the second plot is a bar plot. Use the data from questions 2 and 4.

---



```python
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 5, 2]

plt.subplot(2, 1, 1)
plt.plot(x, y)
plt.title('Line Plot')

plt.subplot(2, 1, 2)
plt.bar(categories, values)
plt.title('Bar Plot')

plt.tight_layout()
plt.show()
```

10.	Create a plot with multiple lines. Plot y1 = [1, 2, 3, 4, 5] and y2 = [5, 4, 3, 2, 1] on the same graph.

---



```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [1, 2, 3, 4, 5]
y2 = [5, 4, 3, 2, 1]

plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.title('Multiple Lines Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

17.	Create a plot with x = [0, 1, 2, 3, 4] and y1 = [0, 1, 4, 9, 16], y2 = [0, 1, 2, 3, 4]. Add a legend for the two lines.

---

20. Create a histogram with two datasets: data1 = [1, 2, 2, 3, 3, 3,

4, 4, 4, 4] and data2 = [1, 3, 3, 3, 4, 4, 4, 5, 5, 6].


---



```python
import matplotlib.pyplot as plt

data1 = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
data2 = [1, 3, 3, 3, 4, 4, 4, 5, 5, 6]

plt.hist([data1, data2], bins=5, label=['Data1', 'Data2'])
plt.title('Histogram with Multiple Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
