1.	Plot a simple line graph where the x-values range from 0 to 10 and the y-values are the squares of the x-values. Label the x-axis as 'x' and the y-axis as 'y = x^2'. Give the plot a title of 'Simple Line Plot'

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 11)
y = x ** 2

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y = x^2')
plt.title('Simple Line Plot')
plt.show()
```

Create a 3D surface plot for z = sin(x) * cos(y)

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(x) * np.cos(y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')
plt.title('3D Surface Plot')
plt.show()
```

25.	Create a pie chart of any dataset and explode the largest slice, add a shadow, and set an alpha value of 0.7.

```python
import matplotlib.pyplot as plt

# Sample data
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]

# Find the largest slice
explode = [0.1 if size == max(sizes) else 0 for size in sizes]

# Define colors with transparency (alpha)
colors = [(1, 0, 0, 0.7), (0, 1, 0, 0.7), (0, 0, 1, 0.7), (1, 1, 0, 0.7)]

# Plot the pie chart
plt.figure(figsize=(7, 7))
plt.pie(sizes, explode=explode, labels=labels, shadow=True, startangle=140, autopct='%1.1f%%', colors=colors)

# Set aspect ratio to be equal so that pie is drawn as a circle
plt.axis('equal')

plt.show()

```

27.	Create a multiple bar plot showing the number of hours spent on three activities (Study, Exercise, Sleep) by three different individuals (Person A, Person B, Person C). Use the data: Person A - [4, 1, 7], Person B - [5, 2, 6], Person C - [3, 1, 8]. Display the bars side by side and use different colors for each person.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
activities = ['Study', 'Exercise', 'Sleep']
person_a = [4, 1, 7]
person_b = [5, 2, 6]
person_c = [3, 1, 8]

# X-axis positions
x = np.arange(len(activities))

# Width of bars
width = 0.2

# Plotting
plt.bar(x - width, person_a, width, label='Person A', color='b')
plt.bar(x, person_b, width, label='Person B', color='g')
plt.bar(x + width, person_c, width, label='Person C', color='r')

# Labeling
plt.xlabel('Activities')
plt.ylabel('Hours Spent')
plt.title('Hours Spent on Activities by Different Individuals')
plt.xticks(x, activities)
plt.legend()

plt.show()
```

30.	Create a 2x2 grid of subplots. In the first subplot (top-left), plot a line graph of y=x2. In the second subplot (top-right), plot a scatter plot of x and y where y is a random sample from a normal distribution. In the third subplot (bottom-left), plot a bar chart of categories A, B, and C with heights 3, 7, and 2, respectively. In the fourth subplot (bottom-right), plot a histogram of 1000 random samples from a normal distribution.

*  **Top-left**: A line graph of y = x ^ 2
*  **Top-right**: A scatter plot of x and y, where y is a random sample from a normal distribution
*  **Bottom-left**: A bar chart of categories A, B, and C with heights 3, 7, and 2, respectively
*  **Bottom-right:** A histogram of 1000 random samples from a normal distribution

```python
import matplotlib.pyplot as plt
import numpy as np

# values for x and y  for line and scatter plot
x = np.linspace(-10, 10, 100)
y = x**2

# For bar graph
categories = ['A', 'B', 'C']
heights = [3, 7, 2]

#random samples from normal distribution for histogram plot
data = np.random.randn(1000)

#ploting according to 2 x 2 Subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

#Top - left
axs[0, 0].plot(x, y, 'r-')
axs[0, 0].set_title('Line Plot')

#Top- Right
axs[0, 1].scatter(x, np.random.randn(len(x)))
axs[0, 1].set_title('Scatter Plot')

#Bottom - Left
axs[1, 0].bar(categories, heights)
axs[1, 0].set_title('Bar Chart')

#Bottom - Right
axs[1, 1].hist(data, bins=30)
axs[1, 1].set_title('Histogram')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(-10, 10, 100)
y1 = x**2
x_scatter = np.linspace(-10, 10, 100)
y_scatter = np.random.normal(size=100)
categories = ['A', 'B', 'C']
heights = [3, 7, 2]
hist_data = np.random.normal(size=1000)

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Top-left subplot
axs[0, 0].plot(x, y1)
axs[0, 0].set_title('y = x^2')

# Top-right subplot
axs[0, 1].scatter(x_scatter, y_scatter)
axs[0, 1].set_title('Scatter Plot')

# Bottom-left subplot
axs[1, 0].bar(categories, heights)
axs[1, 0].set_title('Bar Chart')

# Bottom-right subplot
axs[1, 1].hist(hist_data, bins=30)
axs[1, 1].set_title('Histogram')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
